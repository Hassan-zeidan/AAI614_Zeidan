Question 1: Identify all the outliers in the above data.  Justify your answers using objective measures.

```python
df = grads.copy()
numeric_columns = df.select_dtypes(include=np.number).columns         # picking only the numeric column so we can find the outliers
Q1 = df[numeric_columns].quantile(0.25)                               # calculating the lower quartile
Q2 = df[numeric_columns].quantile(0.75)                               # calculating the upper quartile
IQR = Q2-Q1                                                           # calculating the inter quartile range
lower_bound = Q1-1.5*IQR                                               # calculating the lower bound
upper_bound = Q2+1.5*IQR                                               # calculating the upper bound
outliers = (df[numeric_columns] < lower_bound) | (df[numeric_columns] > upper_bound)
rows_with_any = outliers.any(axis=1)
outlier_rows = df.loc[rows_with_any].copy()
outlier_rows
```

Question 2: There are various data that are missing.  Fill-in the missing data or delete the rows/columns that you think you should delete.  Justify your answer

In the delete part nothing is deleted since there is no column with more than 40% missing values or NaN , same thing applies for the rows , nothing missses more than 60% of the values .. now for the missing values i divided them into two parts , catagories and numerical , for the catagories I filled them with the most repeated one  "most frequent" while the numerical colms I filled them with the mean value of the column .

```python
from sklearn.impute import SimpleImputer

df = grads.copy()

# Drop mostly missing columns/rows
COL_THRESH = 0.40  # >40% missing -> drop column
ROW_THRESH = 0.60  # >60% missing -> drop row
miss_frac = df.isna().mean()                     # fraction missing per column
to_drop  = miss_frac[miss_frac > COL_THRESH].index
df = df.drop(columns=to_drop)
row_miss = df.isna().mean(axis=1)                 # fraction missing per row
df = df.loc[row_miss <= ROW_THRESH].copy()

# Impute ( identify catag and num col)
num_cols = df.select_dtypes(include='number').columns
cat_cols = df.select_dtypes(exclude='number').columns

if len(num_cols):
    df[num_cols] = SimpleImputer(strategy='mean').fit_transform(df[num_cols])
if len(cat_cols):
    df[cat_cols] = SimpleImputer(strategy='most_frequent').fit_transform(df[cat_cols])

grads_clean_q2 = df

grads_clean_q2
```

Question 3: Reload the data and fill-in the data using mean method as well as the frequent method.

```python
import pandas as pd
from sklearn.impute import SimpleImputer

grads2 = pd.read_csv("https://raw.githubusercontent.com/harmanani/AAI614/main/Week%203/grads.csv")

num_cols = grads2.select_dtypes(include='number').columns
cat_cols = grads2.select_dtypes(exclude='number').columns

if len(num_cols):
    grads2[num_cols] = SimpleImputer(strategy='mean').fit_transform(grads2[num_cols])
if len(cat_cols):
    grads2[cat_cols] = SimpleImputer(strategy='most_frequent').fit_transform(grads2[cat_cols])

grads_clean_q3 = grads2
grads_clean_q3
```

Exercise II. Run the cell below to create a new dataframe called `df_miss`.  Its first column will contain some missing values.

```python
import pandas as pd
import numpy as np
import random

nrows = 10
ncols = 5

# set a seed for random number generation
np.random.seed(314)
# create an array filled with random data
data = np.array(np.random.rand(nrows, ncols))
# put the data to a pandas dataframe
df_miss = pd.DataFrame(data)
# rename the columns
df_miss.columns = ['col_'+str(ii) for ii in range(ncols)]

# randomly set some values to missing
ix0 = np.random.randint(nrows, size=3)
ix1 = np.random.randint(nrows, size=3)

df_miss['col_0'][ix0] = np.nan
df_miss['col_1'][ix1] = np.nan

print(df_miss)
```

Impute the missing values (NaN) in `col_0` (but not `col_1`) with the median.  Store the values in the dataframe by using the parameter `inplace`.  Print the dataframe.

```python
df_miss['col_0'].fillna(df_miss['col_0'].median(), inplace=True)
df_miss
```

Impute the missing values in `col_1` with value 0.  Store the values in the dataframe by using the parameter `inplace`.  Print the dataframe.

```python
df_miss['col_1'].fillna(0,inplace=True)
df_miss
```
